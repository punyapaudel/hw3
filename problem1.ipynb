# Part a

## Euler Method

```python
import math
# Initial conditions
x = 0
y = 0

# Step size (h)
h = 0.1

# How many steps to take
steps = 5

print(f"Start: x = {x}, y = {y}, y_exact = {math.tan(x)}, error = {0}")

# Loop to perform Euler's Method
for i in range(steps):
    # 1. Calculate the slope: f(x, y) = y^2 + 1
    slope = y**2 + 1
    
    # 2. Update y: y_new = y_old + h * slope
    y = y + (h * slope)
    
    # 3. Update x: x_new = x_old + h
    x = x + h

    # 4. Exact solution y_exact = tan (x)
    y_exact = math.tan(x)

    # 5. Error (Difference)
    error = abs(y_exact - y)
    
    print(f"Step {i+1}: x = {round(x, 2)}, y = {y:.5f}, y_exact = {y_exact:.5f}, error = {error:.5f}")
```

As $x$ increases, the Euler method (being a first-order method) begins to lag behind the actual curve because it assumes the slope is constant across the entire interval $h$.

## Runge-Kutta Method

```python
import math
# Initial conditions
x = 0
y = 0

# Step size (h)
h = 0.1

# How many steps to take
steps = 5

print(f"Start: x = {x}, y = {y}, y_exact = {math.tan(x)}, error = {0}")

# Loop to perform RK2 (Heun's Method)
for i in range(steps):
    # 1. Calculate k1 (Slope at the start)
    # This is exactly the same as the Euler slope: k1 = f(x, y) = y^2 + 1
    k1 = y**2 + 1
    
    # 2. Predict y at the end of the interval (Euler step)
    # We need this temporary y to find the slope at the end
    y_end_predict = y + h * k1
    
    # 3. Calculate k2 (Slope at the end): 
    # We use the x at the end (x + h) and the predicted y: k2 = f(x+h, y_end_predict)
    k2 = y_end_predict**2 + 1
    
    # 4. Update y: Average of start slope (k1) and end slope (k2)
    y = y + (h / 2) * (k1 + k2)
    
    # 5. Update x: x_new = x_old + h
    x = x + h
    
    # 6. Exact solution y_exact = tan(x)
    y_exact = math.tan(x)
    
    # 7. Error (Difference)
    error = abs(y_exact - y)
    
    print(f"Step {i+1}: x = {round(x, 2)}, y = {y:.5f}, y_exact = {y_exact:.5f}, error = {error:.5f}")
```

## Comparision

Using the same step size ($h=0.1$) and number of steps ($5$), the Runge-Kutta (RK2) method is significantly more accurate than the Euler method. At $x=0.5$, the Euler method has an error of approximately 0.015, while the RK2 method has an error of only 0.0007, demonstrating the superior convergence of higher-order methods.

```python
import matplotlib.pyplot as plt
import math

# EULER
x_eplot = [0]       # Start with initial x
y_eplot = [0]       # Start with initial y
y_exact_plot = [0] # Start with initial exact value

xe = 0
ye = 0

# Step size (h)
h = 0.1

# How many steps to take
steps = 15

# Loop to perform Euler's Method
for i in range(steps):
    # 1. Calculate the slope: f(x, y) = y^2 + 1
    slope = ye**2 + 1
    
    # 2. Update y: y_new = y_old + h * slope
    ye = ye + (h * slope)
    
    # 3. Update x: x_new = x_old + h
    xe = xe + h

    # 4. Exact solution y_exact = tan (x)
    y_exact = math.tan(xe)

    # 5. Error (Difference)
    error = abs(y_exact - ye)
    

    # Store the values
    x_eplot.append(xe)
    y_eplot.append(ye)
    y_exact_plot.append(y_exact)

# RK2
x_rplot = [0]
y_rplot = [0]
xr = 0
yr = 0


# Loop to perform RK2 (Heun's Method)
for i in range(steps):
    # 1. Calculate k1 (Slope at the start)
    # This is exactly the same as the Euler slope: k1 = f(x, y) = y^2 + 1
    k1 = yr**2 + 1
    
    # 2. Predict y at the end of the interval (Euler step)
    # We need this temporary y to find the slope at the end
    y_end_predict = yr + h * k1
    
    # 3. Calculate k2 (Slope at the end): 
    # We use the x at the end (x + h) and the predicted y: k2 = f(x+h, y_end_predict)
    k2 = y_end_predict**2 + 1
    
    # 4. Update y: Average of start slope (k1) and end slope (k2)
    yr = yr + (h / 2) * (k1 + k2)
    
    # 5. Update x: x_new = x_old + h
    xr = xr + h
    
    # 6. Exact solution y_exact = tan(x)
    y_exact = math.tan(xr)
    
    # 7. Error (Difference)
    error = abs(y_exact - yr)

    x_rplot.append(xr)
    y_rplot.append(yr)
    
    

plt.plot(x_eplot, y_eplot, 'bo--', label='Euler Method (h = 0.1)')
plt.plot(x_rplot, y_rplot, 'ko--', label='RK2 Method (h = 0.1)')
plt.plot(x_eplot, y_exact_plot, 'r-', label='Exact Solution (tan x)')

plt.title("Euler Method, RK2 Method and Exact Solution")
plt.xlabel("x")
plt.ylabel("y")
#plt.ylim(-5, 13)
plt.legend()
plt.grid(True)
```

As the value of $x$ gets close to $\pi/2 (\approx 1.57)$, the numerical solution starts to break down.\\

The analytical solution is $y = \tan(x)$. The tangent function has a vertical asymptote at $x = \pi/2$. So, at this point, $y$ goes to infinity.\\

Since $dy/dx = y^2 + 1$, as $y \to \infty$, the slope (derivative) becomes infinitely steep.

Lets calculate the error of both methods at the value of $x = 0.5$ with the step size of $h = 0.1$ and $h = 0.01$.

```python
import math

target_x = 0.5

# The "True" answer at x = 0.5
exact_y = math.tan(target_x)

print(f"Exact Y at x=0.5: {exact_y:.6f}\n")


# PART 1: Check Euler Method

print("Euler Method")

# Test 1: Big Step (h = 0.1)
y = 0
h = 0.1
steps = int(target_x / h) # 5 steps

for i in range(steps):
    slope = y**2 + 1
    y = y + h * slope

error_big = abs(exact_y - y)
print(f"Error with h=0.1:  {error_big:.6f} (y = {y:.6f})")

# Test 2: Small Step (h = 0.01)
y = 0
h = 0.01
steps = int(target_x / h) # 50 steps

for i in range(steps):
    slope = y**2 + 1
    y = y + h * slope

error_small = abs(exact_y - y)
print(f"Error with h=0.01: {error_small:.6f} (y = {y:.6f})")

# Calculate how much better it got
print(f"Improvement: {error_big / error_small:.1f} times better\n")


# PART 2: Check RK2 Method

print("Runge-Kutta (RK2) Method")

# Test 1: Big Step (h = 0.1)
y = 0
h = 0.1
steps = int(target_x / h)

for i in range(steps):
    k1 = y**2 + 1
    y_pred = y + h * k1
    k2 = y_pred**2 + 1
    y = y + (h/2) * (k1 + k2)

error_big = abs(exact_y - y)
print(f"Error with h=0.1:  {error_big:.6f} (y = {y:.6f})")

# Test 2: Small Step (h = 0.01)
y = 0
h = 0.01
steps = int(target_x / h)

for i in range(steps):
    k1 = y**2 + 1
    y_pred = y + h * k1
    k2 = y_pred**2 + 1
    y = y + (h/2) * (k1 + k2)

error_small = abs(exact_y - y)
print(f"Error with h=0.01: {error_small:.8f} (y = {y:.6f})")

# Calculate how much better it got
print(f"Improvement: {error_big / error_small:.1f} times better")
```

Decreasing the step size increases accuracy for both methods. However, the Runge-Kutta method improves much faster. Decreasing the step size by a factor of 10 reduced the Euler error by a factor of ~10, but it reduced the RK2 error by a factor of ~100.

# Part d

```python
import math
import matplotlib.pyplot as plt

def get_final_y(method, h, target_x):
    """
    Runs integration (Euler or RK2) from x=0 to target_x
    and returns the final y value.
    """
    x = 0
    y = 0
    steps = round(target_x / h)
    
    for _ in range(steps):
        if method == "Euler":
            slope = y**2 + 1
            y = y + h * slope
        elif method == "RK2":
            k1 = y**2 + 1
            y_pred = y + h * k1
            k2 = y_pred**2 + 1
            y = y + (h/2) * (k1 + k2)
        x += h
        
    return y

# 1. SETUP
target_x = 1.0  # A safe point

# Define our "Truth" (Highest Resolution Case)
# We use a tiny step size to approximate the "real" answer
h_best = 0.00001
y_best_euler = get_final_y("Euler", h_best, target_x)
y_best_rk2   = get_final_y("RK2",   h_best, target_x)

# Exact solution
exact_y = math.tan(target_x)

print(f"Baseline (High-Res, h = {h_best}) Euler at x = {target_x}: {y_best_euler:.8f}")
print(f"Baseline (High-Res, h = {h_best}) RK2   at x = {target_x}: {y_best_rk2:.8f}")
print(f"Exact solution at x = {target_x}: {exact_y:.8f}\n")

# 2. CONVERGENCE STUDY
# We will test these larger step sizes
#h_values = [0.1, 0.05, 0.02, 0.01, 0.005, 0.002, 0.001]
h_values = [0.1, 0.01, 0.001, 0.0001]

euler_y_values = []
rk2_y_values = []
euler_diffs = []
rk2_diffs = []

print(f"{'h':<8} |    {'Euler y'}      | {'Euler Diff':<15} |      {'RK2 y'}      |{'RK2 Diff':<15}")
print("-" * 80)

for h in h_values:
    # Run Euler
    y_euler = get_final_y("Euler", h, target_x)
    euler_y_values.append(y_euler)
    # Calculate Fractional Difference vs Best Euler
    diff_e = abs(y_euler - y_best_euler) / abs(y_best_euler)
    euler_diffs.append(diff_e)
    
    # Run RK2
    y_rk2 = get_final_y("RK2", h, target_x)
    rk2_y_values.append(y_rk2)
    # Calculate Fractional Difference vs Best RK2
    diff_r = abs(y_rk2 - y_best_rk2) / abs(y_best_rk2)
    rk2_diffs.append(diff_r)
    
    print(f"{h:<8} | {y_euler:<15.8f} | {diff_e:<15.8f} | {y_rk2:<15.8} | {diff_r:<15.8f}")

# --- 3. PLOT RESULTS (Log-Log) ---
plt.figure(figsize=(8, 6))

# Plot Euler
plt.loglog(h_values, euler_diffs, 'bo-', label='Euler Method')

# Plot RK2
plt.loglog(h_values, rk2_diffs, 'rs-', label='RK2 Method')

# Add reference slopes (optional, but helps visualization)
# Just visual guides to show what slope 1 and 2 look like
plt.loglog(h_values, h_values, 'k:', alpha=0.3, label='Reference Slope 1')
plt.loglog(h_values, [h**2 for h in h_values], 'k--', alpha=0.3, label='Reference Slope 2')

plt.xlabel('Step Size (h)')
plt.ylabel('Fractional Difference (vs High-Res)')
plt.title('Convergence Study: Error vs Step Size (Log-Log)')
plt.grid(True, which="both", ls="-", alpha=0.4)
plt.legend()
plt.show()
```

On a log-log plot, the data points form straight lines. This indicates a "power law" relationship ($Error \propto h^n$).\\

Euler Method: The slope is approximately 1. This confirms it is a 1st-order method ($Error \propto h^1$). If you decrease $h$ by 10, the error drops by 10.RK2 Method: The slope is approximately 2. This confirms it is a 2nd-order method ($Error \propto h^2$). If you decrease $h$ by 10, the error drops by 100 ($10^2$).\\

The steeper slope of the RK2 line proves that it converges to the true answer much faster than the Euler method as the step size is reduced.
